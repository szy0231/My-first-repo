# let's do a value strategy:
# basically means that to use a metric of stocks as the standard to choose whether or not to buy a stock.

let's do some preparation work first:

```python
#library import
import numpy as np
import pandas as pd
import math
import yfinance as yf
```

```python
# get a tickers' batch:
tickers=pd.read_html("https://en.wikipedia.org/wiki/List_of_S%26P_500_companies")[0]['Symbol']
tickers=tickers.to_list()
tickers
```

```python
# Let's try only one metric for one stock first
apple=yf.Ticker('AAPL')
print(apple.info)
```

```python
# make batch calls
chunk_size=150
first_result=[]
for i in range(0,len(tickers),chunk_size):
    chunk_tickers=tickers[i:i+chunk_size]
    chunk_tickers_str=' '.join(chunk_tickers)
    #make one batch call
    chunk_stocks=yf.Tickers(chunk_tickers_str)
    print(chunk_stocks)
    for ticker in chunk_tickers:
        single_stock_info=chunk_stocks.tickers[ticker].info
        price=single_stock_info.get('regularMarketOpen')
        P_over_E=single_stock_info.get('trailingPE')
        first_result.append(
            {
                'Ticker':ticker,
                "Price":price,
                "P/E":P_over_E,
                'Number of Share to Buy':'N/A'
            }
        )


```

```python
first_result=pd.DataFrame(first_result)
final_result=first_result.dropna(subset=['Price']).dropna(subset=['P/E']).reset_index(drop=True)
final_result.sort_values('P/E',ascending=False,inplace=True)
final_result=final_result[0:50].reset_index(drop=True)
final_result
```

```python
# build our own position size:
def put_your_money():
    global portfolio_size
    portfolio_size=input('how much money you will put here, boy?')
    try:
        float(portfolio_size)
    except ValueError:
        print("Don't fuck with it!")
        portfolio_size=input('how much money you will put here, boy?')

put_your_money()
position_size=float(portfolio_size)/len(final_result)
position_size
```

```python
#calculate shares we should buy (let's say for each stock in the first 50 P/E values, we buy the equal share)
for i in final_result.index:
    final_result.loc[i,'Number of Share to Buy']=position_size/final_result.loc[i,'Price']

final_result
```

# Plus Altra!!!!
# we not only use one metric but also others, therefore we should use an average score measure to represent a comprehensive value metrics
# we will use P/E, P/B (price-to-book), and EV/EBITDA(Enterprise value Mutiples), which are considered as the less the more value for a stock

```python
# make batch calls
chunk_size=150
first_result=[]
for i in range(0,len(tickers),chunk_size):
    chunk_tickers=tickers[i:i+chunk_size]
    chunk_tickers_str=' '.join(chunk_tickers)
    #make one batch call
    chunk_stocks=yf.Tickers(chunk_tickers_str)
    print(chunk_stocks)
    for ticker in chunk_tickers:
        single_stock_info=chunk_stocks.tickers[ticker].info
        price=single_stock_info.get('regularMarketOpen')
        P_over_E=single_stock_info.get('forwardPE')
        P_B=single_stock_info.get('priceToBook')
        ev_to_ebitda=single_stock_info.get('enterpriseToEbitda')

        first_result.append(
            {
                'Ticker':ticker,
                "Price":price,
                "P/E":P_over_E,
                "P/E Percentile":'N/A',
                "Price_to_book":P_B,
                "Price_to_book Percentile":'N/A',
                "Ev_to_Ebitda":ev_to_ebitda,
                "Ev_to_Ebitda Percentile":'N/A',
                'RV Score':'N/A',
                'Number of Share to Buy':'N/A'
            }
        )
first_result=pd.DataFrame(first_result)
first_result
```

```python
#refine the results
final_result=first_result.dropna(subset=['Price']).dropna(subset=['P/E']).dropna(subset=['Price_to_book']).dropna(subset=['Ev_to_Ebitda']).reset_index(drop=True)
#calculate the percentile
Percentile_forename=['P/E','Price_to_book','Ev_to_Ebitda']
for i in Percentile_forename:
    final_result[f'{i} Percentile']=final_result[f'{i}'].rank(pct=True)

final_result
```

```python
# Calculate the RV Score
for i in final_result.index:
    final_result.loc[[i],['RV Score']]=final_result.iloc[i,[3,5,7]].mean()

#Selecting the best 50 stocks (with lower value of each three metrics --- lower RV Score, the better)
final_result=final_result.sort_values('RV Score',ascending=True)
Best_value_stocks=final_result.iloc[0:50].reset_index(drop=True)
Best_value_stocks
```

```python
#calculate shares we should buy (let's say for each stock in the first 50 P/E values, we buy the equal share)
for i in Best_value_stocks.index:
    Best_value_stocks.loc[i,'Number of Share to Buy']=position_size/Best_value_stocks.loc[i,'Price']

Best_value_stocks
```
